```python
from bs4 import BeautifulSoup
import requests
import csv
```

```python
key = 'toys' # search keyword
url = 'https://www.ebay.com/sch/i.html?_from=R40&_nkw={}&_sacat=246&_pgn='.format(key) # url with keyword format

datas = [] # empty dataframe
for page in range(0, 10): # number of page
  req = requests.get(url+str(page)) # request web page based on methods
  soup = BeautifulSoup(req.text, 'html.parser') #beautiful soup parser
  items = soup.findAll('div', 's-item__wrapper clearfix')

  # formulate csv data
  for it in items:
    Name = it.find('h3', 's-item__title').text.replace('New Listing', '')
  
    try : Category = it.find('span', 'SECONDARY_INFO').text
    except : Category = 'Other'
 
    try : Location = it.find('span', 's-item__location s-item__itemLocation').text
    except : Location = 'Other'
  
    try : Discount = it.find('span', 's-item__discount s-item__discount').find('span', 'BOLD').text
    except : Discount = 'No Discount'

    try : Price = it.find('div', 's-item__detail s-item__detail--primary').find('span', 's-item__price').text.replace('$','').split(' ')[0]
    except : Price = '0'
  
    datas.append([Name, Category, Location, Discount, Price])
```

```python
table_headers = ['Name', 'Category', 'Location', 'Discount', 'Price'] # create header
writer = csv.writer(open('sample_data/ebay_{}.csv'.format(key), 'w', newline='')) # write parameter

# fill row with page content
writer.writerow(table_headers)
for d in datas: writer.writerow(d)
```
